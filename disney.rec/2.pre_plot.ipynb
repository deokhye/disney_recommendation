```
#영어, 특수문자 지우기
import re

def clean_text(text):
    cleaned_text = re.sub('[a-zA-z]','',text)
    cleaned_text = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"\♥\♡\ㅋ\ㅠ\ㅜ\ㄱ\ㅎ\ㄲ\ㅡ]1234567890','',cleaned_text)
    return cleaned_text
```

```
import pandas as pd
import numpy as np
import csv

from tensorflow.keras.preprocessing.text import Tokenizer
from sklearn.model_selection import train_test_split
```

```
data = pd.read_csv("movies_plot.csv", index_col=False, encoding='utf-8', names=['code', 'plot'])    #datafram으로 바꿔오기
#print(data)

#nan 값 제거
data['plot'] = data['plot'].str.replace("[^\w]", " ")
data['plot'] = data['plot'].replace('', np.nan)
data['plot'] = data['plot'].replace('개봉 후 평점은 개봉당일 오전 9시 이후부터 입력가능합니다 ', np.nan)
data = data.dropna(how='any')
data = data.reset_index()
data = data.iloc[:,1:]
#print(data)

#csv로 저장
data.to_csv('plot_processing.csv')  
print(data)



# for i in data['plot']:
#     print(i)
#     words = Okt().morphs(data[i])

# #stopwords 지정
#     sw = ['를', '을', '이', '가', '으로', '로', '에', '에서', 'ABOUT', 'THE', 'MOVIE', '1', '2', '3', '4', 'HOT', 'ISSUE', "", ".", ",", "(", ")", "?", "!", "<", ">", "&", "PRODUCTION", "NOTE", "5", "#", "%" ]

#     sw_removed = []
#     if i.lower() not in sw:
#         sw_removed.append(i)
```

```
plot = data['plot']
lst = []

for i in plot:
    lst.append(clean_text(i))
#print(lst)

data['plot'] = lst
#print(data)

data.to_csv("pre_plot.csv", index=False)
print(data)
```

```
# print(data)
# print(len(data['code']))

# print(data['code'])

#코드 리스트
code_lst = []
for i in data['code']:
    code_lst.append(i)
# print(code_lst)

#중복되는 코드 찾기
dup_code = []
for i in range(len(code_lst)-1):
    if code_lst[i] == code_lst[i+1]:
        dup_code.append(code_lst[i])
#print(dup_code)

#코드를 index로
data_index = data
data_index = data_index.set_index('code', drop = True, append=False, inplace=False)
#print(data_index)

#중복되는 코드면 plot을 리스트에 넣기
lst=[]
for i in data['code']:
    if i in dup_code:
        lst.append(data_index.loc[i, 'plot'])
#print(lst)

csv_lst = []
#리스트로 변환
for i in range(len(lst)//2): 
    csv_lst.append(list(np.array(lst[i*2].tolist())))

plot_lst = []
#두 리스트  하나로 합치기
for i in range(len(csv_lst)):
    new_line = csv_lst[i][0] + csv_lst[i][1]
    plot_lst.append(new_line)
print(plot_lst)


#딕셔너리 형태로 만들기
dup_dic = {}

for i in range(len(dup_code)):
    code = dup_code[i]
    plot = plot_lst[i]
    dup_dic[code] = plot
#print(dup_dic)

#중복 안된 데이터
not_dup_code = []
for i in data['code']:
    if i not in dup_code:
        not_dup_code.append(i)
# print(not_dup_code)
# print(dup_code)

#plot을 리스트에 넣기
not_lst=[]
for i in data['code']:
    if i in not_dup_code:
        not_lst.append(data_index.loc[i, 'plot'])
#print(not_lst)

not_plot_lst = []
#리스트로 변환
for i in not_lst: 
    not_plot_lst.append([i])
#print(not_plot_lst)

#딕셔너리 형태로 만들기
not_dup_dic = {}

for i in range(len(not_dup_code)):
    code = not_dup_code[i]
    plot = not_plot_lst[i]
    not_dup_dic[code] = plot
#print(not_dup_dic)

#두개 딕셔너리 합치고 csv로 전환
total_dict = {**dup_dic, **not_dup_dic}
print(total_dict)

with open('fianl_plot.csv', 'w', encoding='utf-8') as f:
    for key in total_dict.keys():
        f.write("%s,%s\n"%(key,total_dict[key]))
        
```
